## Importing libraries

## Defining the types of variables

```python
from investor import Investor
```

## Defining the System Params

```python
from typing import List

investor1 = Investor("Alice", .4, 4, 16)
investor2 = Investor("Bob", .2, 12, 48)

investors = (investor1,investor2)
        
@dataclass
class Parameters:
    investors: List[Investor]
    initial_token_supply: int = 1000000

# Initialize Parameters instance with default values
system_params = Parameters(investors).__dict__
```

## State Variables

```python

@dataclass
class StateVariables:
        investors = [Investor]
        circulating_token_supply: int = 0
        months: int = 0

initial_state = StateVariables().__dict__
```

## Policy Functions

```python
def p_timestep(params, substep, state_history, prev_state, **kwargs):
    months = prev_state['months']
    months += 1
    return {'months': months}


def p_vesting_schedule(params, substep, state_history, prev_state, **kwargs):
    
    months = prev_state['months']
    
    if months == 0:
        investors = params['investors']
    else:   
        investors = prev_state['investors']
        
    circulating_token_supply = prev_state['circulating_token_supply']
    
    for investor in investors:
        if months < investor.get_cliff_months():
            continue
        
        tokens_to_allocate = params['initial_token_supply'] * investor.vesting_percent / investor.issuing_months
        circulating_token_supply -= tokens_to_allocate
        
        investor.allocated_tokens += tokens_to_allocate
    
    return {'investors': investors, 'circulating_token_supply': circulating_token_supply}

```

## State Update Functions and Partial State Update Blocks

```python

def s_months(params, substep, state_history, prev_state,policy_input, **kwargs):
    '''Update the state of the circulating_token_supply'''

    updated_months = np.ceil(prev_state['months'] + policy_input['months'])
    return ('months', updated_months)




def s_circulating_token_supply(params, substep, state_history, prev_state,policy_input, **kwargs):
    '''Update the state of the circulating_token_supply'''

    updated_circulating_token_supply = np.ceil(prev_state['circulating_token_supply'] + policy_input['circulating_token_supply'])
    return ('circulating_token_supply', updated_circulating_token_supply)


def s_investors(params, substep, state_history, prev_state,policy_input, **kwargs):
    '''Update the state of the circulating_token_supply'''

    updated_investors = policy_input['investors']
    return ('investors', updated_investors)





state_update_blocks = [
    {
        'policies': {
            'p_vesting_schedule': p_vesting_schedule,
            'p_timestep':p_timestep

        },
        'variables': {
            'months': s_months,
            'circulating_token_supply': s_circulating_token_supply,
            'investors': s_investors

        }
    },

]

```

```python
# config and run

#number of timesteps
TIMESTEPS = 40
#number of monte carlo runs
RUNS = 5


model = Model(initial_state=initial_state, state_update_blocks=state_update_blocks, params=system_params)
simulation = Simulation(model=model, timesteps=TIMESTEPS, runs=RUNS)

experiment = Experiment(simulation)
# Select the Pathos backend to avoid issues with multiprocessing and Jupyter Notebooks
experiment.engine = Engine(backend=Backend.PATHOS)

result = experiment.run()


df = pd.DataFrame(result)
df
```

## Scenario Analysis

Let's say the game team has a shop in the game where certain items cost 50 coins. And they believe that the median player should have to play 10 games to earn these items to adequently incentivize them. What should be the initial value of the crash chance parameter to get 5 as the median coins

```python
TIMESTEPS = 30
RUNS = 100

model = Model(initial_state=initial_state, state_update_blocks=state_update_blocks, params=system_params)
simulation = Simulation(model=model, timesteps=TIMESTEPS, runs=RUNS)

def median_coin_100_runs(i):
    simulation.model.params.update({
        # Running a parameter sweep on the initial chance of crashing 
        'crash_chance':[i]
    })
    experiment = Experiment(simulation)

    # Select the Pathos backend to avoid issues with multiprocessing and Jupyter Notebooks
    experiment.engine = Engine(backend=Backend.PATHOS)

    result = experiment.run()


    df = pd.DataFrame(result)

    median_coins = df['coins'].median()
    return median_coins

```

```python
median_coins_expected = 5
crash_chance_initial_guess = 1
step_distance = 3


guess = crash_chance_initial_guess

for i in range(20):
    median_for_guess = median_coin_100_runs(guess)
    print(i,', crash chance = ',guess, ', median_for_guess = ', median_for_guess)
    if median_for_guess>median_coins_expected:
        guess+=step_distance
    else:
        break

print(guess)
```

Therefore to get a median coin value of 5 we need crash_chance parameter = 22
